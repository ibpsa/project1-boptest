# Weather uncertanty emulator
Notebook to simulate the different uncertainty scenarios by using the function to generate the error.

```python
from ErrorEmulator import predict_error
import numpy as np
import matplotlib.pyplot as plt
```

Low uncertainty scenario

```python
hp = 48
F0_low = 0
K0_low = 0.6
F_low = 0.92
K_low = 0.4
mu_low = 0
Nsim = 50000
e_low = np.zeros((Nsim,hp))
for i in range(Nsim):
    e_low[i,:] = predict_error(hp, F0_low, K0_low, F_low, K_low, mu_low)
```

```python
mean_e_low = e_low.mean(axis=0)
std_e_low = e_low.std(axis=0)
plt.plot(mean_e_low, label='Mean error')
plt.plot(std_e_low, label='Std error')
plt.xlabel('Prediction horizon [h]')
plt.ylabel('Temperature [ºC]')
plt.legend()
plt.title('Low uncertainty scenario')
```

Intermediate uncertainty scenario - zero mean

```python
F0_int1 = 0.15
K0_int1 = 1.2
F_int1 = 0.93
K_int1 = 0.6
mu_int1 = 0
e_int1 = np.zeros((Nsim,hp))
for i in range(Nsim):
    e_int1[i,:] = predict_error(hp, F0_int1, K0_int1, F_int1, K_int1, mu_int1)
```

```python
F0_int2 = 0.3
K0_int2 = 1
F_int2 = 0.93
K_int2 = 0.5
mu_int2 = 0.015
e_int2 = np.zeros((Nsim,hp))
for i in range(Nsim):
    e_int2[i,:] = predict_error(hp, F0_int2, K0_int2, F_int2, K_int2, mu_int2)
```

High uncertainty scenario

```python
F0_high = -0.58
K0_high = 1.5
F_high = 0.95
K_high = 0.7
mu_high = -0.015
e_high = np.zeros((Nsim,hp))
for i in range(Nsim):
    e_high[i,:] = predict_error(hp, F0_high, K0_high, F_high, K_high, mu_high)
```

Comparison of the scenarios

```python
tp = range(1,49)
color_low = [0.4660, 0.6740, 0.1880]
color_int1 = [0.1529, 0.6078, 0.9098]
color_int2 = [0, 0.4470, 0.7410]
color_high = [0.6350, 0.0780, 0.1840]
plt.plot(tp,mean_e_low, label='Low uncertainty', color=color_low)
plt.plot(tp,mean_e_int1, label='Intermediate uncertainty - zero mean', color=color_int1)
plt.plot(tp,mean_e_int2, label='Intermediate uncertinty - nonzero mean', color=color_int2)
plt.plot(tp,mean_e_high, label='High uncertainty', color=color_high)
plt.xlabel('Prediction horizon [h]')
plt.ylabel('Mean temperature error [ºC]')
plt.title('Mean error')
plt.legend(bbox_to_anchor=(1.05,1))
```
